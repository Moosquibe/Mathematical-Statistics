### Example Wasserman 9.14.4 Page 147

Want to show that the MLE for $\theta$ is consistent, where $X_1, X_2, ... , X_n$ $\sim$ $Uniform(0,\theta$)

The MLE of $\theta$ is $Y = max(X_1, X_2, ... , X_n)$ 

*Consistent* means that the MLE converges in probability to the true value

i.e. $P(|Y - \theta| > \epsilon) \rightarrow 0 $ as $ n  \rightarrow \infty$

Recall $P(Y \leq c) = P(X_1 \leq c, X_2 \leq c, ... , X_n \leq c) = P(X_1 \leq c) P(X_2 \leq c) ... P(X_n \leq c)$

Consider



$
P(Y \leq t ) = \Bigg \{
\begin{array}{c}
 0 & if & t \leq 0\\\
(\frac{t}{\theta})^n & if & 0 < t < \theta  \\\
 1 & if & t \geq \theta 
\end{array}
$

For $\epsilon > 0$

$P(|Y - \theta| > \epsilon)= P(Y \geq   \theta + \epsilon) + P(Y \leq   \theta - \epsilon)$
$
 = \Big \{
\begin{array}{c}
(\frac{\theta - \epsilon }{\theta})^n & for & \epsilon < \theta  \\\
 0 & for & \epsilon \geq \theta 
\end{array}
$ $\to 0$ as $n \to \infty$

### Example CB 7.65

**(a)** Plotting the LINEX loss given by:

$$ L(\theta, a) = e^{c(a-\theta)} - c(a-\theta) - 1$$


For $c = 0.2, 0.5, 1$ and $\theta = 1$ and $10$ 


```python
theta = np.array([.0, 2])
C = np.array([0.2, 0.5, 1])
a = np.linspace(-1,2.5,num=200)
col = ['blue', 'red']

plt.figure(figsize=(12,6))
for i, th in enumerate(theta):
    for c in C:
        plt.plot(a, np.exp(c*(a-th)) - c * (a - th) - 1, c = col[i], alpha = c, label = '$\Theta$ = {}, a = {}'.format(th,c))
        plt.ylim(-0.01,0.4)
plt.title('Plot of $L(\Theta, a)$ for $\Theta$ = {} ({}), and {} ({})\n'.format(theta[0], col[0], theta[1],col[1])
          , fontsize = 18)
plt.xlabel('a', fontsize = 16)
plt.ylabel('$L(\Theta, a)$', fontsize = 16)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.legend(fontsize = 14);
```
